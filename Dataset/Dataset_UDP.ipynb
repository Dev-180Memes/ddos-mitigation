```python
import numpy as np
import pandas as pd
udp = pd.read_csv("01-12/DrDoS_UDP.csv")
udp.head()
```

```python
udp1 = udp.drop([' Source IP', ' Source Port', ' Destination IP', ' Destination Port', ' Timestamp', 'SimillarHTTP'], axis=1)
udp1.dropna()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

udp1[' Label'] = label_encoder.fit_transform(udp1[' Label'])
udp1[' Label'].unique()
```

```python
udp1.to_csv("01-12-Clean/UDP_Clean.csv")
```
