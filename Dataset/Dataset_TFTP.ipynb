```python
import numpy as np
import pandas as pd

tftp = pd.read_csv("01-12/TFTP.csv")
tftp.head()
```

```python
tftp1 = tftp.drop([' Source IP', ' Source Port', ' Destination IP', ' Destination Port', ' Timestamp', 'SimillarHTTP'], axis=1)
tftp1.dropna()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

tftp1[' Label'] = label_encoder.fit_transform(tftp1[' Label'])
tftp1[' Label'].unique()
```

```python
tftp1.to_csv("01-12-Clean/TFTP_Clean.csv")
```
