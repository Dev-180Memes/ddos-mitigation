```python
# extract 60% of the dataframe
df = df.sample(frac=0.6, random_state=30)
```

```python
df.drop(['index'], inplace=True, axis=1)
```

```python
from sklearn.model_selection import train_test_split

features = [' Flow Duration', ' Flow IAT Mean', ' Flow IAT Std', ' Flow IAT Max', ' Flow IAT Min', 'Fwd IAT Total', ' Fwd IAT Mean', ' Fwd IAT Std', ' Fwd IAT Max', ' Fwd IAT Min', 'Bwd IAT Total', ' Bwd IAT Mean', ' Bwd IAT Std', ' Bwd IAT Max', ' Bwd IAT Min']

target = [' Label']

X = df.loc[:, features]
y = df.loc[:, target]
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=30)
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
pred_clf = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score
```

```python
accuracy_score(y_test, pred_clf)
```

```python
precision_score(y_test, pred_clf)
```

```python
f1_score(y_test, pred_clf)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier()
```

```python
ada.fit(X_train, y_train)
```

```python
pred_ada = ada.predict(X_test)
```

```python
accuracy_score(y_test, pred_ada)
```

```python
precision_score(y_test, pred_ada)
```

```python
f1_score(y_test, pred_ada)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import StackingClassifier
```

```python
models = [('clf', RandomForestClassifier()), ('ada', AdaBoostClassifier())]

meta = LogisticRegression()
```

```python
stack = StackingClassifier(estimators=models, final_estimator=meta)
```

```python
stack.fit(X_train, y_train)
```

```python
pred = stack.predict(X_test)
```

```python
accuracy_score(y_test, pred)
```

```python
precision_score(y_test, pred)
```
